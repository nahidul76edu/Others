```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

w = 0.19
X = ['1s', '2s', '3s']

#Online_ICT_Express order X1=accuracy, X2=Precision,Recall,F1 score,Cohens kappa 

X1 = [88.21, 91.95, 92.26]
X2 = [92.98, 93.30, 92.57]
X3 = [84.10, 91.27, 92.86]
X4 = [87.05, 92.12, 92.51]
X5 = [76.56, 84.56, 84.50]

bar1 = np.arange(len(X))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]
bar5 = [i+w for i in bar4]

plt.figure(figsize=(12,7))

fontP = FontProperties()
fontP.set_size('large')

font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 14}

plt.rc('font', **font)
#plt.axhspan(0, 1, facecolor='white', alpha=0.5)
plt.rcParams['figure.facecolor'] = 'white'
plt.ylim(0, 100)

p1= plt.bar(bar1, X1, w, label = 'Accuracy')
p2 = plt.bar(bar2, X2, w, label = 'Precision')
p3 = plt.bar(bar3, X3, w, label = 'Recall')
p4 = plt.bar(bar4, X4, w, label = 'F1 score')
p5 = plt.bar(bar5, X5, w, label = 'Cohens kappa')

plt.xticks(bar1+0.375, X)

#plt.legend()
plt.legend(handles=[p1, p2, p3, p4, p5], title='Performance', bbox_to_anchor=(0.995, 1.017), loc='upper left', prop=fontP)

#plt.style.use(['classic'])
#plt.savefig('Performance_sub1to4_dpi200.png', dpi = 300)
#plt.savefig(('Performance_sub1to4.png'), dpi=200, format='png', bbox_inches='tight')
#plt.show()


def insert_data_labels(bars):
	for bar in bars:
		bar_height = bar.get_height()
		plt.annotate('{0:.2f}'.format(bar.get_height()),
			xy=(bar.get_x() + bar.get_width() / 2, bar_height),
			xytext=(0, 3),
			textcoords='offset points',
			ha='center',
			va='bottom'
		)

insert_data_labels(p1)
insert_data_labels(p2)
insert_data_labels(p3)
insert_data_labels(p4)
insert_data_labels(p5)

plt.ylabel("Performance")

plt.savefig(('KNN & CKNN.png'), dpi=300, format='png', bbox_inches='tight')
plt.show()
```
