```python
def raw_plot(data):
  plt.figure(figsize=(7,5))
  plt.suptitle('Raw Data', fontsize=16)
  plt.xlabel('common xlabel')
  plt.ylabel('common ylabel')

  font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 3}
  plt.rc('font', **font)

  line_wide=0.5

  plt.subplot(5,1,1)
  plt.plot(data[0],linewidth=line_wide)
  plt.gca().xaxis.set_major_locator(pyplot.NullLocator())

  plt.subplot(5,1,2)
  plt.plot(data[1],linewidth=line_wide)
  plt.gca().xaxis.set_major_locator(pyplot.NullLocator())

  plt.subplot(5,1,3)
  plt.plot(data[2],linewidth=line_wide)
  plt.gca().xaxis.set_major_locator(pyplot.NullLocator())

  plt.subplot(5,1,4)
  plt.plot(data[3],linewidth=line_wide)
  plt.gca().xaxis.set_major_locator(pyplot.NullLocator())

  plt.subplot(5,1,5)
  plt.plot(data[4],linewidth=line_wide)
  
  re_size=224
  plt.savefig('AEP RAW DATA.png', dpi=200)

  plt.show()
```

```python
def one_signal(data,number):
  i=number*256
  class1=data.iloc[i:i+256]
  return class1
```
