```python
import tensorboard as tf
import keras
```

```python
input = keras.Input(shape=(100,), dtype='int32', name='input')
x = keras.layers.Embedding(output_dim=512, input_dim=10000, input_length=100)(input)
x = keras.layers.LSTM(32)(x)
x = keras.layers.Dense(64, activation='relu')(x)
x = keras.layers.Dense(64, activation='relu')(x)
x = keras.layers.Dense(64, activation='relu')(x)
output = keras.layers.Dense(1, activation='sigmoid', name='output')(x)
model = keras.Model(inputs=[input], outputs=[output])
dot_img_file = '/tmp/model_1.png'#keras.utils.plot_model(model, to_file=dot_img_file, show_shapes=True)
```
